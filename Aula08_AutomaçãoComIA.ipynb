```python
# Este script verifica a temperatura de um sistema e alerta se estiver muito alta.

# Solicita ao usuário que insira a temperatura atual.
# A função input() retorna uma string, então precisamos convertê-la para um número (float).
while True:
    try:
        temperature_input = input("Por favor, insira a temperatura atual em °C: ")
        temperature = float(temperature_input)
        break # Sai do loop se a conversão for bem-sucedida
    except ValueError:
        print("Entrada inválida. Por favor, insira um número para a temperatura.")

# Exibe a temperatura que foi inserida pelo usuário.
# Agora usando concatenação para formatar a saída, mostrando a temperatura com duas casas decimais.
print("Temperatura atual inserida: " + "{:.2f}".format(temperature) + "°C")

# Define o limite de temperatura para alerta.
THRESHOLD = 30.0 # Em graus Celsius

# Verifica se a temperatura ultrapassa o limite definido.
# Se a temperatura for maior que o THRESHOLD, um alerta é exibido.
if temperature > THRESHOLD:
    print("ALERTA: Temperatura de aquecimento detectada! Tome medidas.\n")
    print("Dica: Verifique os ventiladores, o fluxo de ar e a carga do sistema.")
# Caso contrário, informa que a temperatura está dentro dos limites normais.
else:
    print("Temperatura normal. O sistema está operando dentro dos limites seguros.")
```

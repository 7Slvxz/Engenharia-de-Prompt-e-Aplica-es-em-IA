```python
# Este script verifica a temperatura de um sistema e alerta se estiver muito alta.

def get_temperature_input():
    """
    Solicita ao usuário que insira a temperatura atual e garante que a entrada seja um número.
    Retorna a temperatura como um float.
    """
    while True:
        try:
            temperature_input = input("Por favor, insira a temperatura atual em °C: ")
            temperature = float(temperature_input)
            return temperature
        except ValueError:
            print("Entrada inválida. Por favor, insira um número para a temperatura.")

def check_temperature_and_alert(temperature, threshold):
    """
    Verifica se a temperatura ultrapassa o limite definido e exibe o alerta apropriado.
    Args:
        temperature (float): A temperatura atual a ser verificada.
        threshold (float): O limite de temperatura para acionar o alerta.
    """
    print("Temperatura atual inserida: " + "{:.2f}".format(temperature) + "°C")

    if temperature > threshold:
        print("ALERTA: Temperatura de aquecimento detectada! Tome medidas.\n")
        print("Dica: Verifique os ventiladores, o fluxo de ar e a carga do sistema.")
    else:
        print("Temperatura normal. O sistema está operando dentro dos limites seguros.")

def main():
    """
    Função principal que orquestra a obtenção da temperatura e a verificação do alerta.
    """
    # Define o limite de temperatura para alerta.
    THRESHOLD = 30.0 # Em graus Celsius

    current_temperature = get_temperature_input()
    check_temperature_and_alert(current_temperature, THRESHOLD)

# Bloco principal para executar o script quando ele é chamado diretamente.
if __name__ == '__main__':
    main()
```

```python
import os
import shutil

def organize_files(source_directory):
    """
    Organiza arquivos em um diretório específico em subpastas
    baseadas na extensão do arquivo.

    Args:
        source_directory (str): O caminho do diretório a ser organizado.
    """

    # Dicionário para mapear extensões para nomes de pastas
    file_types = {
        'Imagens': ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'],
        'Documentos': ['.pdf', '.doc', '.docx', '.txt', '.odt', '.rtf', '.xls', '.xlsx', '.ppt', '.pptx'],
        'Áudios': ['.mp3', '.wav', '.aac', '.flac', '.ogg'],
        'Vídeos': ['.mp4', '.mov', '.avi', '.mkv', '.flv', '.wmv'],
        'Compactados': ['.zip', '.rar', '.7z', '.tar', '.gz'],
        'Código': ['.py', '.java', '.c', '.cpp', '.html', '.css', '.js'],
        'Executáveis': ['.exe', '.msi', '.deb', '.rpm'],
        'Outros': [] # Para arquivos com extensões não mapeadas
    }

    # Cria as pastas de destino se não existirem
    for folder_name in file_types.keys():
        destination_path = os.path.join(source_directory, folder_name)
        if not os.path.exists(destination_path):
            os.makedirs(destination_path)
            print(f"Pasta '{folder_name}' criada.")

    # Itera sobre os arquivos no diretório de origem
    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Ignora diretórios e as pastas recém-criadas
        # A condição `filename == os.path.basename(__file__)` foi removida
        # porque '__file__' não é definido em ambientes de notebook
        # e o script não está no diretório que está sendo organizado.
        if os.path.isdir(file_path) or filename in file_types.keys():
            continue

        file_extension = os.path.splitext(filename)[1].lower() # Pega a extensão e converte para minúsculas

        moved = False
        for folder, extensions in file_types.items():
            if file_extension in extensions:
                destination_folder = os.path.join(source_directory, folder)
                # Para evitar o erro 'Destination path already exists', podemos verificar antes de mover
                # ou, de forma mais robusta, limpar o diretório de exemplo antes de cada execução
                # Já que vamos implementar a limpeza no bloco __main__, podemos manter o shutil.move direto aqui.
                shutil.move(file_path, destination_folder)
                print(f"Movido '{filename}' para '{folder}'")
                moved = True
                break

        # Se a extensão não foi encontrada em nenhuma categoria, move para 'Outros'
        if not moved:
            destination_folder = os.path.join(source_directory, 'Outros')
            shutil.move(file_path, destination_folder)
            print(f"Movido '{filename}' para 'Outros'")

    print("Organização de arquivos concluída!")

if __name__ == '__main__':
    # Define um diretório temporário para os exemplos
    temp_dir = 'arquivos_para_organizar_exemplo'

    # Limpa o diretório de exemplo antes de cada execução para garantir um estado limpo
    if os.path.exists(temp_dir):
        print(f"Limpando o diretório existente '{temp_dir}'...")
        shutil.rmtree(temp_dir)
        print(f"Diretório '{temp_dir}' e seu conteúdo removidos.")

    # Cria o diretório de exemplo após a limpeza
    os.makedirs(temp_dir)
    print(f"Diretório de exemplo '{temp_dir}' criado.")

    # Cria 6 arquivos de exemplo no diretório temporário
    example_files = [
        'documento_importante.pdf',
        'foto_ferias.jpg',
        'musica_favorita.mp3',
        'video_aula.mp4',
        'backup_dados.zip',
        'relatorio.log' # Um arquivo para 'Outros'
    ]

    print(f"\nCriando {len(example_files)} arquivos de exemplo em '{temp_dir}'...")
    for filename in example_files:
        with open(os.path.join(temp_dir, filename), 'w') as f:
            f.write(f'Conteúdo de {filename}')
        print(f"Criado: {filename}")

    # Define o diretório a ser organizado para o diretório temporário
    source_dir_to_organize = temp_dir
    print(f"\nIniciando organização de arquivos em '{source_dir_to_organize}'...")
    organize_files(source_dir_to_organize)

    # Opcional: Limpar o diretório de exemplo após a execução
    # Para manter os resultados visíveis após a execução, deixaremos essa parte comentada.
    # Se quiser limpar automaticamente, descomente as linhas abaixo.
    # print(f"\nLimpando o diretório de exemplo '{temp_dir}'...")
    # if os.path.exists(temp_dir):
    #     shutil.rmtree(temp_dir)
    #     print(f"Diretório '{temp_dir}' e seu conteúdo removidos.")
```

```python
import requests # Importa a biblioteca requests para fazer requisições HTTP a APIs externas.

def buscar_endereco(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço a partir de um CEP.
    Inclui tratamento de exceções para lidar com possíveis erros na requisição ou na resposta.
    """
    # Constrói a URL da API ViaCEP com o CEP fornecido.
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        # Realiza a requisição GET à API ViaCEP.
        response = requests.get(url)
        # Levanta um HTTPError para respostas de status ruins (códigos 4xx ou 5xx).
        response.raise_for_status()
        # Converte a resposta JSON em um dicionário Python.
        data = response.json()

        # Verifica se a resposta JSON contém um campo 'erro', indicando que o CEP não foi encontrado.
        if "erro" in data:
            print(f"CEP {cep} não encontrado. Verifique o número digitado.")
            return None
        else:
            # Se o CEP for encontrado, imprime as informações do endereço formatadas.
            print(f"\nInformações para o CEP {cep}:")
            # Usa .get() para evitar KeyError caso alguma chave não exista na resposta da API.
            print(f"  Rua: {data.get('logradouro', 'N/A')}")
            print(f"  Bairro: {data.get('bairro', 'N/A')}")
            print(f"  Cidade: {data.get('localidade', 'N/A')}")
            print(f"  Estado: {data.get('uf', 'N/A')}")
            print(f"  DDD: {data.get('ddd', 'N/A')}")
            # Retorna o dicionário com os dados do endereço.
            return data
    except requests.exceptions.RequestException as e:
        # Captura erros relacionados à requisição (ex: problemas de conexão, URL inválida).
        print(f"Erro de conexão ou requisição: {e}")
        return None
    except ValueError:
        # Captura erros caso a resposta da API não seja um JSON válido.
        print("Erro ao decodificar a resposta JSON. A API pode ter retornado um formato inesperado.")
        return None
    except Exception as e:
        # Captura quaisquer outros erros inesperados.
        print(f"Ocorreu um erro inesperado: {e}")
        return None

# Bloco principal de execução do script quando ele é executado diretamente.
if __name__ == '__main__':
    # Loop infinito para permitir múltiplas consultas de CEP até o usuário decidir sair.
    while True:
        # Solicita ao usuário que digite um CEP ou 'sair' para encerrar.
        cep_input = input("Digite o CEP (apenas números, ou 'sair' para encerrar): ")
        # Verifica se o usuário deseja sair do programa.
        if cep_input.lower() == 'sair':
            break

        # Valida o formato do CEP: deve conter apenas dígitos e ter 8 caracteres.
        if not cep_input.isdigit() or len(cep_input) != 8:
            print("CEP inválido. Por favor, digite 8 dígitos numéricos.")
            continue # Continua para a próxima iteração do loop, pedindo um novo CEP.

        # Chama a função para buscar o endereço com o CEP fornecido.
        buscar_endereco(cep_input)
```

```python
import time # Importa o módulo time para usar funções relacionadas a tempo, como time.sleep() e time.ctime().
import random # Importa o módulo random para gerar números aleatórios, simulando leituras de sensores.

# --- Refactoring: Variável de controle global para parar o monitoramento ---
# Esta variável booleana controla o estado de execução do sistema de monitoramento.
_global_monitoring_should_run = True

def stop_program_gracefully():
    """
    Define a variável global '_global_monitoring_should_run' como False para indicar
    que o sistema de monitoramento deve parar sua execução de forma controlada.
    Esta função pode ser chamada para 'fechar' o programa de monitoramento.
    """
    global _global_monitoring_should_run
    _global_monitoring_should_run = False
    print("\n--- Sinal de parada recebido. O monitoramento será encerrado em breve. ---")

def check_condition(value, critical_threshold):
    """
    Verifica se o valor monitorado atingiu uma condição crítica.
    Compara o 'value' atual com o 'critical_threshold' para determinar se a condição é crítica.
    Args:
        value (float): O valor atual a ser verificado (e.g., temperatura).
        critical_threshold (float): O limite a partir do qual a condição é considerada crítica.
    Returns:
        bool: True se a condição for crítica (value > critical_threshold), False caso contrário.
    """
    return value > critical_threshold

def send_alert(message, alert_type='console'):
    """
    Simula o envio de um alerta, formatando a saída de acordo com o tipo de alerta.
    Pode exibir o alerta no console ou simular o envio de um e-mail.
    Args:
        message (str): A mensagem de alerta a ser enviada.
        alert_type (str): O tipo de alerta ('console' para imprimir no console ou 'email_simulado' para simular um e-mail).
    """
    if alert_type == 'console':
        print(f"\n!!! ALERTA DE SISTEMA (Console) !!!")
        print(f"Data/Hora: {time.ctime()}") # Exibe a data e hora atuais do alerta.
        print(f"Mensagem: {message}\n")
    elif alert_type == 'email_simulado':
        print(f"\n--- Email Simulado Enviado ---")
        print(f"Para: equipe_monitoramento@empresa.com")
        print(f"Assunto: ALERTA CRÍTICO DE SISTEMA")
        print(f"Corpo: {message}")
        print(f"------------------------------\n")
    else:
        print(f"Tipo de alerta desconhecido: {alert_type}") # Mensagem de erro para tipos de alerta inválidos.

def autonomous_monitoring_system(interval_seconds=5, critical_temp=30.0):
    """
    Sistema autônomo de monitoramento que verifica continuamente uma condição
    e dispara alertas quando a condição crítica é atingida.
    Ele agora verifica uma variável global para determinar se deve continuar rodando.
    Args:
        interval_seconds (int): Intervalo de tempo (em segundos) entre as verificações do sensor.
        critical_temp (float): A temperatura crítica que aciona o alerta.
    """
    global _global_monitoring_should_run # Declara que vamos usar a variável global para controle.
    _global_monitoring_should_run = True # Garante que o monitoramento começa ativo ao ser chamado.

    print(f"Iniciando sistema de monitoramento autônomo. Verificando a cada {interval_seconds} segundos...")
    print(f"Temperatura crítica definida: {critical_temp}°C")

    # --- Refactoring: O loop agora verifica a variável de controle '_global_monitoring_should_run' ---
    while _global_monitoring_should_run: # O loop continua enquanto esta flag for True.
        # Simula a leitura de um sensor (e.g., temperatura) gerando um valor aleatório.
        current_value = random.uniform(20.0, 35.0) # Gera um valor aleatório entre 20 e 35°C.
        print(f"[{time.ctime()}] Valor atual do sensor: {current_value:.2f}°C")

        # Verifica se o valor atual excede a temperatura crítica.
        if check_condition(current_value, critical_temp):
            alert_message = f"Condição CRÍTICA detectada! Temperatura {current_value:.2f}°C excedeu o limite de {critical_temp}°C."
            send_alert(alert_message, alert_type='console') # Envia um alerta para o console.
            # send_alert(alert_message, alert_type='email_simulado') # Descomente para simular o envio de um email.
        else:
            print("Condição normal.") # Informa que a condição está dentro dos limites seguros.

        time.sleep(interval_seconds) # Pausa a execução pelo tempo definido no intervalo.

    print("--- Sistema de monitoramento encerrado. ---")


# Ponto de entrada do programa: O código abaixo é executado apenas quando o script é rodado diretamente.
if __name__ == '__main__':
    # Você pode ajustar os parâmetros aqui para configurar o monitoramento:
    monitoring_interval = 3 # Define o intervalo de verificação para 3 segundos.
    critical_temperature_threshold = 32.0 # Define o limite de 32°C para acionar o alerta.

    # Inicia o sistema de monitoramento autônomo com os parâmetros configurados.
    autonomous_monitoring_system(monitoring_interval, critical_temperature_threshold)

    # Como este é um ambiente de célula única no Colab, a função `stop_program_gracefully()`
    # precisaria ser chamada de um contexto externo ou em um thread separado para interromper
    # o loop 'while _global_monitoring_should_run' de forma programática enquanto ele está ativo.
    # Em um ambiente interativo como o Colab, o método mais comum para parar uma célula em execução
    # com um loop infinito é usar o botão de 'Parar Execução' da célula.
    # No entanto, a função `stop_program_gracefully` agora está disponível para ser usada em
    # um cenário de programa mais complexo (e.g., com múltiplas threads, eventos ou interfaces de usuário)
    # onde o controle de parada é externo ao loop principal.
```

```python
# Este script verifica a temperatura de um sistema e alerta se estiver muito alta.

# Solicita ao usuário que insira a temperatura atual.
# A função input() retorna uma string, então precisamos convertê-la para um número (float).
while True:
    try:
        temperature_input = input("Por favor, insira a temperatura atual em °C: ")
        temperature = float(temperature_input)
        break # Sai do loop se a conversão for bem-sucedida
    except ValueError:
        print("Entrada inválida. Por favor, insira um número para a temperatura.")

# Exibe a temperatura que foi inserida pelo usuário.
# Agora usando concatenação para formatar a saída, mostrando a temperatura com duas casas decimais.
print("Temperatura atual inserida: " + "{:.2f}".format(temperature) + "°C")

# Define o limite de temperatura para alerta.
THRESHOLD = 30.0 # Em graus Celsius

# Verifica se a temperatura ultrapassa o limite definido.
# Se a temperatura for maior que o THRESHOLD, um alerta é exibido.
if temperature > THRESHOLD:
    print("ALERTA: Temperatura de aquecimento detectada! Tome medidas.\n")
    print("Dica: Verifique os ventiladores, o fluxo de ar e a carga do sistema.")
# Caso contrário, informa que a temperatura está dentro dos limites normais.
else:
    print("Temperatura normal. O sistema está operando dentro dos limites seguros.")
```

```python
import os
import shutil

def organize_files(source_directory):
    """
    Organiza arquivos em um diretório específico em subpastas
    baseadas na extensão do arquivo.

    Args:
        source_directory (str): O caminho do diretório a ser organizado.
    """

    # Dicionário para mapear extensões para nomes de pastas
    file_types = {
        'Imagens': ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'],
        'Documentos': ['.pdf', '.doc', '.docx', '.txt', '.odt', '.rtf', '.xls', '.xlsx', '.ppt', '.pptx'],
        'Áudios': ['.mp3', '.wav', '.aac', '.flac', '.ogg'],
        'Vídeos': ['.mp4', '.mov', '.avi', '.mkv', '.flv', '.wmv'],
        'Compactados': ['.zip', '.rar', '.7z', '.tar', '.gz'],
        'Código': ['.py', '.java', '.c', '.cpp', '.html', '.css', '.js'],
        'Executáveis': ['.exe', '.msi', '.deb', '.rpm'],
        'Outros': [] # Para arquivos com extensões não mapeadas
    }

    # Cria as pastas de destino se não existirem
    for folder_name in file_types.keys():
        destination_path = os.path.join(source_directory, folder_name)
        if not os.path.exists(destination_path):
            os.makedirs(destination_path)
            print(f"Pasta '{folder_name}' criada.")

    # Itera sobre os arquivos no diretório de origem
    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Ignora diretórios e as pastas recém-criadas
        # A condição `filename == os.path.basename(__file__)` foi removida
        # porque '__file__' não é definido em ambientes de notebook
        # e o script não está no diretório que está sendo organizado.
        if os.path.isdir(file_path) or filename in file_types.keys():
            continue

        file_extension = os.path.splitext(filename)[1].lower() # Pega a extensão e converte para minúsculas

        moved = False
        for folder, extensions in file_types.items():
            if file_extension in extensions:
                destination_folder = os.path.join(source_directory, folder)
                # Para evitar o erro 'Destination path already exists', podemos verificar antes de mover
                # ou, de forma mais robusta, limpar o diretório de exemplo antes de cada execução
                # Já que vamos implementar a limpeza no bloco __main__, podemos manter o shutil.move direto aqui.
                shutil.move(file_path, destination_folder)
                print(f"Movido '{filename}' para '{folder}'")
                moved = True
                break

        # Se a extensão não foi encontrada em nenhuma categoria, move para 'Outros'
        if not moved:
            destination_folder = os.path.join(source_directory, 'Outros')
            shutil.move(file_path, destination_folder)
            print(f"Movido '{filename}' para 'Outros'")

    print("Organização de arquivos concluída!")

if __name__ == '__main__':
    # Define um diretório temporário para os exemplos
    temp_dir = 'arquivos_para_organizar_exemplo'

    # Limpa o diretório de exemplo antes de cada execução para garantir um estado limpo
    if os.path.exists(temp_dir):
        print(f"Limpando o diretório existente '{temp_dir}'...")
        shutil.rmtree(temp_dir)
        print(f"Diretório '{temp_dir}' e seu conteúdo removidos.")

    # Cria o diretório de exemplo após a limpeza
    os.makedirs(temp_dir)
    print(f"Diretório de exemplo '{temp_dir}' criado.")

    # Cria 6 arquivos de exemplo no diretório temporário
    example_files = [
        'documento_importante.pdf',
        'foto_ferias.jpg',
        'musica_favorita.mp3',
        'video_aula.mp4',
        'backup_dados.zip',
        'relatorio.log' # Um arquivo para 'Outros'
    ]

    print(f"\nCriando {len(example_files)} arquivos de exemplo em '{temp_dir}'...")
    for filename in example_files:
        with open(os.path.join(temp_dir, filename), 'w') as f:
            f.write(f'Conteúdo de {filename}')
        print(f"Criado: {filename}")

    # Define o diretório a ser organizado para o diretório temporário
    source_dir_to_organize = temp_dir
    print(f"\nIniciando organização de arquivos em '{source_dir_to_organize}'...")
    organize_files(source_dir_to_organize)

    # Opcional: Limpar o diretório de exemplo após a execução
    # Para manter os resultados visíveis após a execução, deixaremos essa parte comentada.
    # Se quiser limpar automaticamente, descomente as linhas abaixo.
    # print(f"\nLimpando o diretório de exemplo '{temp_dir}'...")
    # if os.path.exists(temp_dir):
    #     shutil.rmtree(temp_dir)
    #     print(f"Diretório '{temp_dir}' e seu conteúdo removidos.")
```

Prompt 1

Voce é um Dev em Python Senior, faça uma calculadora simples, com menu de interaçao com opçao de saida (break).

```python
def calculadora():
    while True:
        print("\n--- CALCULADORA PYTHON ---")
        print("1. Soma (+)")
        print("2. Subtração (-)")
        print("3. Multiplicação (*)")
        print("4. Divisão (/)")
        print("5. Sair")

        opcao = input("\nEscolha uma opção (1-5): ")

        if opcao == '5':
            print("Encerrando o programa... Até logo!")
            break

        if opcao in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Digite o primeiro número: "))
                num2 = float(input("Digite o segundo número: "))

                if opcao == '1':
                    print(f"Resultado: {num1 + num2}")
                elif opcao == '2':
                    print(f"Resultado: {num1 - num2}")
                elif opcao == '3':
                    print(f"Resultado: {num1 * num2}")
                elif opcao == '4':
                    if num2 == 0:
                        print("Erro: Não é possível dividir por zero!")
                    else:
                        print(f"Resultado: {num1 / num2}")
            except ValueError:
                print("Erro: Por favor, digite apenas números.")
        else:
            print("Opção inválida! Tente novamente.")

if __name__ == "__main__":
    calculadora()
```

Prompt 2

Refatore esse codigo e utilize o minimo de linhas para melhorar o processamento da maquina

```python
import operator

def calc():
    ops = {'1': ('+', operator.add), '2': ('-', operator.sub), '3': ('*', operator.mul), '4': ('/', operator.truediv)}
    while (opt := input("\n1:+ 2:- 3:* 4:/ 5:Sair\n> ")) != '5':
        if opt in ops:
            try:
                n1, n2 = float(input("N1: ")), float(input("N2: "))
                print(f"Res: {ops[opt][1](n1, n2)}")
            except ZeroDivisionError: print("Erro: Divisão por zero")
            except ValueError: print("Erro: Apenas números")
        else: print("Inválido")

if __name__ == "__main__":
    calc()
```
